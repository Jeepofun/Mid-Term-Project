```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('PassengersRawData.csv')
```

```python
df = df[df['dest_country'] == df['origin_country']]
```

```python
df['ground_time'] = df['ramp_to_ramp'] - df['air_time']
df['capacity_used'] = df['passengers']/df['seats']
df['estimated_passenger_weight'] = df['passengers'] * 210
df['estimated_payload_used'] = (df['estimated_passenger_weight'] + df['freight'] + df['mail'])/df['payload']
df['cancelation_rate'] = ((df['departures_performed']/df['departures_scheduled'])-1)*-1
```

```python
df.replace([np.inf, -np.inf], 0, inplace = True)
df[df['cancelation_rate']<0]['cancelation_rate'].replace(df[df['cancelation_rate']<0]['cancelation_rate'], 0)
df['cancelation_rate'] = df['cancelation_rate']*1
```

```python
#sorted(list(df.columns))
```

```python

df.fillna(0, inplace=True)
df.isnull().values.any()
```

```python
df
```

```python
df = df[['unique_carrier','airline_id','origin_airport_id','dest_airport_id','distance','aircraft_type','class','departures_scheduled','departures_performed','cancelation_rate','air_time','ground_time','ramp_to_ramp',
         'payload','estimated_payload_used','freight','mail', 'seats', 'passengers', 'estimated_passenger_weight','capacity_used','month','year']]
df = df.drop(df[df['departures_performed'] > 0][df['air_time']== 0][df['ground_time'] == 0].index, axis = 0)
```

```python
l = []
for i in df:
    if df[i].dtype != 'object':
        l.append(i)
l
df2 = df [l]
```
