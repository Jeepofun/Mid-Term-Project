# load stuff

```python
import pandas as pd
```

```python

# load all the csv files with the ranks of stuff


dest_ranks = pd.read_csv('dest_ranks.csv')
origin_ranks = pd.read_csv('origin_ranks.csv')
tail_num_ranks = pd.read_csv('tail_num_rank.csv')
carrier_ranks = pd.read_csv('carrier_ranks.csv')
fuel = pd.read_csv('Carrier_Fconsumption_2018.csv')
```

```python
test_data = pd.read_csv('test_data.txt')     # load the test data and take a quick look 
test_data
```

# merge features to the test data

```python
test_data_Wranks = pd.merge(test_data,dest_ranks, on=['dest'], how='left')           
```

```python
test_data_Wranks = pd.merge(test_data_Wranks,origin_ranks, on=['origin'], how='left')
```

```python
test_data_Wranks = pd.merge(test_data_Wranks,tail_num_ranks, on=['tail_num'], how='left')
```

```python
test_data_Wranks = pd.merge(test_data_Wranks,carrier_ranks, on=['mkt_unique_carrier'], how='left')
```

```python
test_data_Wranks.dropna(inplace=True)
```

```python
test_data_Wranks.isnull().sum()
```

```python
fuel['mkt_unique_carrier'] = fuel['unique_carrier']
```

```python
fuel = fuel[['mkt_unique_carrier','sdomt_gallons','tdomt_gallons','ts_cost','tdomt_cost']]
```

```python
test_data_Wranks = pd.merge(test_data_Wranks,fuel, on=['mkt_unique_carrier'], how='left')
```

# create day column

```python
from datetime import datetime

test_data_Wranks['human_time'] = test_data_Wranks['fl_date']/1000
```

```python
test_data_Wranks.head()
```

```python
test_data_Wranks['date'] = test_data_Wranks['human_time'].apply(lambda x: datetime.utcfromtimestamp(x).strftime('%Y-%m-%d %H:%M:%S'))
```

```python
test_data_Wranks['day'] = (pd.to_datetime(test_data_Wranks['human_time'], format='%Y-%m-%d %H:%M:%S')).dt.day
```

```python
test_data_Wranks.head()
```

```python
test_data_Wranks.drop(columns=['tail_num_ranks','carrier_ranks','origin_ranks','dest_ranks'],inplace=True)
```
